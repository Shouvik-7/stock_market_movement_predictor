```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
```

```python
tokenizer = AutoTokenizer.from_pretrained('ProsusAI/finbert')
model = AutoModelForSequenceClassification.from_pretrained('ProsusAI/finbert')
```

```python
from transformers import pipeline
classifier = pipeline('sentiment-analysis', model=model, tokenizer=tokenizer)
```

```python
tweets = df_tweets['Tweet'].to_list()
```

```python
tweets
```

```python
res = []
re_score = []
p = 0
for t in tweets:
    res.append(classifier(t)[0]['label'])
    re_score.append(classifier(t)[0]['score'])
    if p%100 == 0:
        print(p)
    p = p+1
```

```python
import pickle
with open('sentli.pkl', 'wb') as f:
  pickle.dump(res, f)
```

```python
import pickle
with open('sentsc.pkl', 'wb') as f:
  pickle.dump(re_score, f)
```

```python
df_tweets['sentiment'] = res
df_tweets['sentiment_score'] = re_score
```

```python
df1 = df1.set_index('Date')
```

```python
def convert_sentiment(x):
  if x == 'negative':
    return -1
  elif x=='positive':
    return 1
  else:
    return 0
df1['sentiment'] = df1['sentiment'].apply(convert_sentiment)
```

```python
ag_df = df1.groupby([pd.Grouper(freq='1D')])['sentiment'].mean()
ag_df
```

```python
import yfinance as yf
import pandas as pd

# Define the stock symbol and timeframe
stock_symbol = "AAPL"
start_date = "2021-09-30"
end_date = "2022-09-29"

# Fetch stock data using yfinance
stock_data = yf.download(stock_symbol, start=start_date, end=end_date)

# Display the stock data
print(stock_data.head())
```

```python
stock_data
```

```python
merged_df = pd.merge(stock_data, ag_df, left_index=True, right_index=True)
merged_df
```

```python
merged_df['PriceDiff'] =  merged_df['Adj Close'].diff()
merged_df.dropna(inplace=True)
merged_df
```

```python
BEAR = 0
BULL = 1
merged_df['StockMovement'] = np.where(merged_df['PriceDiff'] > 0, BULL, BEAR) 
merged_df
```
